```python
# Group by PropName and find the most common GISPropNum for each group
correct_gispropnum = df.groupby('PropName')['GISPropNum'].agg(lambda x: x.mode().iloc[0]).to_dict()

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Check if PropName is not null
    if pd.notnull(row['PropName']):
        # Get the corresponding correct GISPropNum from the dictionary
        correct_gis = correct_gispropnum.get(row['PropName'])
        # Update the GISPropNum if it's not correct
        if row['GISPropNum'] != correct_gis:
            df.at[index, 'GISPropNum'] = correct_gis
```

```python
def update_omppropid_column(row):
    gis = row['GISPropNum']
    omp = row['OMPPropID']
    
    # If the OMPPropID column contains a '-' and the code before '-' is different, update the value
    if '-' in omp:
        gis_part, second_part = omp.split('-', 1)
        if gis_part != gis:
            return gis + '-' + second_part
    # If the OMPPropID column does not contain '-', update the value if it's different from the GISPropNum column
    elif omp != gis:
        return gis
    
    # Otherwise, return the original value
    return omp

# Apply the custom function to update values in the OMPPropID column
df['OMPPropID'] = df.apply(update_omppropid_column, axis=1)
```

```python
df
```

```python
import numpy as np
def update_borough_column(row):
    gispropnum = row['GISPropNum']
    borough = row['Borough']
    
    # Check if gispropnum is a string (not NaN)
    if isinstance(gispropnum, str):
        # Extract the first letter from the gispropnum in GISPropNum column
        first_letter = gispropnum[0]
        
        # If the borough value in Borough column is different from the first letter of the gispropnum, update it
        if borough != first_letter:
            return first_letter
    # Handle NaN values by returning the original borough value
    return borough

# Apply the custom function to update values in the Borough column
df['Borough'] = df.apply(update_borough_column, axis=1)
```

```python
df
```

```python
def update_department_column(row):
    department = row['Department']
    borough = row['Borough']
    
    borough_part, second_part = department.split('-', 1)
    if borough_part != borough:
        return borough + '-' + second_part
    
    # Otherwise, return the original value
    return department

# Apply the custom function to update values in the OMPPropID column
df['Department'] = df.apply(update_department_column, axis=1)
```
